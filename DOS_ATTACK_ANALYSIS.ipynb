```
import pandas as pd
data=pd.read_csv("DOS attack.csv")
```

```
data.head()
```

```
source=data["Source"]
unique_source=source.value_counts()
print(unique_source)
values=unique_source.to_dict()
print(type(unique_source))
print(values)
keys=values.keys()
vals=values.values()
import matplotlib.pyplot as plt
plt.figure(figsize=(30,20))
plt.bar(keys,vals)
```

```
destination=data["Destination"]
unique_destination=destination.value_counts()
print(unique_destination)
destination_values=unique_destination.to_dict()
print(type(unique_destination))
print(destination_values)
keys=destination_values.keys()
vals=destination_values.values()
import matplotlib.pyplot as plt
plt.figure(figsize=(30,20))
plt.bar(keys,vals)
```

```
info=data["Info"]
data["Findings"]=data["Info"].str.find("SYN")
data["Port"]=data["Info"].str.find("135")
#Protocols In data exchange
protocols=data["Protocol"]
values=protocols.value_counts().to_dict()
keys=values.keys()
vals=values.values()
import matplotlib.pyplot as plt
plt.figure(figsize=(30,20))
plt.bar(keys,vals)
plt.title("Protocol Graph")
#Kind of attack
syn=data["Findings"].value_counts().to_dict()
print(syn)
keys=syn.keys()
vals=syn.values()
import matplotlib.pyplot as plt
plt.figure(figsize=(30,20))
plt.bar(keys,vals)
plt.title("SYN Packets In Information")

#Ports used
values=data["Port"].value_counts().to_dict()
keys=values.keys()
vals=values.values()
import matplotlib.pyplot as plt
plt.figure(figsize=(30,20))
plt.bar(keys,vals)
plt.title("Port Graph")


#ACK TO SYN ATTACKS
data["Findings_ACK"]=data["Info"].str.find("ACK")
values=data["Findings_ACK"].value_counts().to_dict()
keys=values.keys()
vals=values.values()
import matplotlib.pyplot as plt
plt.figure(figsize=(30,20))
plt.bar(keys,vals)
plt.title("ACK TO SYN PACKETS GRAPH")

```
